```python
'''
So Luke says (to paraphrase) “Think about diving into decisions trees. They’re easy to understand 
and provide a great jumping off point for understanding more complex algorithms”. So here we are, 
understanding decision trees (and random forests, since they go together) and working with blobs 
to play with the concept.

It seems that, when thinking first about classification, random forests are a classic example of 
an ensemble learning method. The reason why is because they are an ensemble of many weak learners 
(decision trees). 

To start with decision trees, it seems that the “more modern” term for them is Classification and 
Regression Trees (CART) and they work by using binary splitting to make decisions based on 
features (like “Is Mahogany cool - Yes or No? Ok, how cool is she - Kinda Cool or Amazingly Cool?” 
and so on).

To play more with the algorithm and the concept in general, again, some random data will be 
created using some built-in functions in scikit-learn for bringing up a fake dataset. That starts 
below by putting the make blobs function from the sklearn.datasets to use. X and y will be equal 
to “make_blobs”. The total number of samples will be equal to 500. Having there be 4 centers will 
make it so that there are 125 samples per center. Random state is set to 5 for making it easier to 
share results. If the standard devision is set too low, then the clusters would be completely 
separate and it would be difficult to get a feel for the algorithm. If it is too high of a standard 
devision, there would not be actual clusters (more like blobs in the true sense of the word… 
experience talking). In that situation, it would be hard to tell what the actual results look like. 
So the standard deviation will be set to 2.5.
'''
```

```python
X, y = make_blobs(n_samples = 500, centers = 4,
                  random_state = 5, cluster_std = 2.5)
```

```python
'''
To make a scatterplot below, the figure size will be set to a 12 by 12. Then, it will pull the 0 
and 1 axis from the X information. The color label will be set equal to yellow. The size will 
be set equal to 50. The color map will be set equal to cool.
'''
```

```python
plt.figure(figsize = (12, 12))
plt.scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap = "cool") # "Cool", like Mahogany
```
